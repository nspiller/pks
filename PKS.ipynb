```python
# load packages, define plot settings
import matplotlib.pylab as plt
import numpy as np
plt.rcParams.update({'figure.dpi': 200})
cmap = plt.set_cmap('Paired')
colors = plt.cm.Paired(np.linspace(0,1,12))
x = np.linspace(-2, 2, 100)
spins = [1/2, 3/2, 5/2, 7/2, 9/2]

# PKS energy expression 
def e(j, s, lk, x, b):
    'takes J, spin, lambda^2/k_, x_, B, returns E+, E-'
    a = -j * s * (s + 1) + .5 * lk * x**2
    b = (0.5 * lk**2 * x**2 + b**2 * (s + 0.5)**2 )**.5

    p = a + b
    m = a - b
    return p, m

# plotting function, only pass axis limits
def plot(xmax, ymax):
    fig, ax = plt.subplots(figsize=(5,4))
    for i, s in enumerate(spins):
        yp, ym = e(j, s, lk, x, b)

        ax.plot(x, yp, c=colors[2*i], label='+{}'.format(s))
        ax.plot(x, ym, c=colors[2*i+1], label='-{}'.format(s) )
        
    ax.set_xlabel(r'x_')
    ax.set_ylabel(r'energy [cm-1]')
    ax.set_xlim(-xmax, xmax)
    ax.set_ylim(top=ymax)
    ax.legend(bbox_to_anchor=(1, 1.))
    fig.tight_layout()
```

# Figure 1 from https://doi.org/10.1038/s41557-021-00853-5
